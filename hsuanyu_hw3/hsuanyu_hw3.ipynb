### Problem 1: More Function with Tuples 
Compute sum function, which can take any number of arguments. Reverse the tuple and rotate the tuple. Raise appropriate warning and error when needed.

```python
# Q1-1
def my_sum(*arg):
    output = 0
    for i in arg:
        output = output+i
    return(output)

my_sum(1,2,3)   
```

```python
# Q1-2
def reverse_tuple(tup):
    output = list()
    n = len(tup) - 1
    while n>=0:
        output.append(tup[n])
        n = n-1
    return(tuple(output))

x = (1,2,3,4,5)
reverse_tuple(x)
```

```python
# Q1-3
def rotate_tuple(tup, n):
    try:
        print('The input was not as expected. It should be integer.')
        n = int(n)
    except (TypeError, ValueError):
        raise
    if type(tup) != tuple:
        raise TypeError('The input should be tuple.')
    output = [None]*len(tup)
    l = list(range(0,len(tup)))
    for i in l:
        index = i+(n%len(tup))
        if index >= len(tup):
            output[index-len(tup)] = tup[i]
        else:
            output[index] = tup[i]
    return(output)

tup = (1,2,3,4,5)
n = 3.5
rotate_tuple(tup, n)    
```

### Problem 3: Counting Word Bigrams


```python
# Q3-1
def my_split(s, seprators):
    out = [s]
    for sep in seprators:
        for loop, text in enumerate(out):
            out[loop:loop+1] = [i for i in text.split(sep) if i]
    return out

def count_bigrams_in_file(file):
    import re
    import string
    b1 = open('badstring.txt','r')
    bad = b1.read()
    b1.close()
    st = string.punctuation+bad
    f = open(file, 'r')
    lines = list()
    for i in f:
        x = my_split(i.lower(), st)
        y = ' '.join(x)
        x1 = re.split('\W',y)
        lines.append(x1)
        
    words = [item for sublist in lines for item in sublist if item]
    l = list(range(0,len(words)-1))
    output = {}
    for i in l:
        if (words[i], words[i+1]) not in output:
            output[(words[i], words[i+1])] = 1
        else:
            output[(words[i], words[i+1])] = output[(words[i], words[i+1])] + 1
    return(output)

```

```python
# Q3-2
import os
os.chdir('/hsuanyu_hw3')
import pickle
f = count_bigrams_in_file('mobydick.txt')
output = open('mb.bigrams.pickle','wb')
pickle.dump(f, output)
output.close()

pkl_file = open('mb.bigrams.pickle','rb')
r = pickle.load(pkl_file)
print(r)
```

```python
# Q3-3
def collocations(file):
    col = count_bigrams_in_file(file)
    output = {}
    for (a,b) in col.keys():
        if a not in output:
            output[a] = {b}
        if a in output:
            output[a].add(b)
        if b not in output:
            output[b] = {a}
        if b in output:
            output[b].add(a)
    return(output)

collocations('file.txt')
```
